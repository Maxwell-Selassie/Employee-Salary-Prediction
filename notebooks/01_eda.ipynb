`Data Quality Checks`

```python
# missing values
missing = df.isnull().sum()
missing = missing[missing > 0].sort_values(ascending=False)
missing_pct = (missing / len(df)) * 100
if missing.sum() == 0:
    print({'has missing': False, 'missing' : []})
else:
    summary_df = pd.DataFrame(
        {
            'missing': missing,
            'missing_pct': missing_pct
        }
    )
    print(summary_df)
```

```python
# duplicates check
duplicates = df.duplicated().sum()
if duplicates == 0:
    print({'has_duplicates' : False, 'duplicates': []})
else:
    df[df.duplicated()]
```

```python
# outlier detection using IQR
def outlier_detection(df, col):
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]

    return lower_bound, upper_bound, outliers

for i, col in enumerate(numeric_cols, 1):
    lower, upper, outliers = outlier_detection(df, col)
    print(f'{i:<2}. {col:<20} | Outliers: {len(outliers):<5} | Range: ({lower:.2f} - {upper:.2f})')
```
